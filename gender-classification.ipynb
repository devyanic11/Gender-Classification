```python
X = data.drop(columns=["Gender"])
y = data["Gender"]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
# Initialize encoders
ohe = OneHotEncoder(sparse=False)
le = LabelEncoder()

X_train_transformed = ohe.fit_transform(X)
y_train_transformed = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train_transformed, y_train_transformed, test_size=0.2)
```

```python
y_train_transformed.ndim
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(random_state=0)
```

```python
dt.fit(X_train, y_train)
```

```python
dt_pred = dt.predict(X_test)
```

```python
accuracy_score(y_test, dt_pred)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
```

```python
accuracy_score(y_test, rf_pred)
```

# SVM

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
accuracy_score(y_test, svm_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
accuracy_score(y_test, knn_pred)
```
